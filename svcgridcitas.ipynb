```python
# Agrupar por 'genero' y contar la cantidad de 'decisiones' para cada género
grouped = df.groupby('genero')['decision'].value_counts().unstack(fill_value=0)

# Calcular el total de cada género
totals = grouped.sum(axis=1)

# Calcular la proporción de 'decisiones' para cada género
proportions = grouped.divide(totals, axis=0)

# Crear el gráfico de barras apilado
proportions.plot(kind='bar', stacked=True)

# Personalizar el gráfico
plt.title('Distribución de decisiones por género')
plt.xlabel('Género')
plt.ylabel('Proporción de decisiones')

# Mostrar el gráfico
plt.legend(title='Decision', loc='lower right')
plt.show()
#Aqui vemos que el genero 0 es mas rechazado
```
